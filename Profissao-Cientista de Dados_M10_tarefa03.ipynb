````python
# Importacao de bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Carregamento da base de dados
df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
# Conversao da coluna data_ref para datetime
df['data_ref'] = pd.to_datetime(df['data_ref'])
df.head()
````

````python
df['data_ref'].value_counts()
````

````python
# Novo Dataframe com as variaveis qualitativas
variaveis_qualitativas = df[['data_ref','sexo','posse_de_veiculo','posse_de_imovel','tipo_renda','educacao','estado_civil','tipo_residencia','mau']]
variaveis_qualitativas.head()
````

````python
# Grafico de barras com distribuicao da variavel SEXO no tempo
ax = sns.countplot(x='data_ref', hue='sexo', data=variaveis_qualitativas)
tick_labs = variaveis_qualitativas['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

labels = ax.set_xticklabels(tick_labs, rotation=45)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
ax = sns.pointplot(x='data_ref', y='renda', hue='sexo',
                  data=df, dodge=True, errorbar=('ci', 95))
tick_labs = df['data_ref'].map(lambda ts: ts.strftime('%m-%Y')).unique()
ax.set_xticks(list(range(df['data_ref'].nunique())))
ax.set_xticklabels(tick_labs, rotation=45);

plt.title('Média de Renda por SEXO ao longo do tempo')
````

No grafico acima **"Média de Renda por SEXO ao longo do tempo"** podemos observar que as médias do sexo Feminino e Masculino não apresentam grande variaçao ao longo do tempo, portanto essa variável seria uma boa candidata para modelagem.

````python
ax = sns.pointplot(x='data_ref', y='renda', hue='posse_de_veiculo',
                  data=df, dodge=True, errorbar=('ci', 95))
tick_labs = df['data_ref'].map(lambda ts: ts.strftime('%m-%Y')).unique()
ax.set_xticks(list(range(df['data_ref'].nunique())))
ax.set_xticklabels(tick_labs, rotation=45);

plt.title('Média de Renda por POSSE DE VEICULO ao longo do tempo')
````
